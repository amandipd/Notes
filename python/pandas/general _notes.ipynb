```python
#loc is row, iloc is column

coffee.loc[[0, 1, 2]] # first 3 rows
coffee.loc[[0, 1, 2]] # frst row, 2second row, 6th row
coffee.loc[5:] # 6th row and onwards

coffee.loc[:, ["Day", "Units Sold"]]
coffee.iloc[:,[0,2]]

coffee.loc[1, "Units:Sold"] = 10 # changes Units:Sold of first row to 10

coffee["Day"] # accessing Day column
coffee.sort_values("Units Sold") # increasing order
coffee.sort_values("Units Sold", "Coffee Type"], ascending=False) # decreasing order

# Looping through rows in dataframe

for index, row in coffee.iterrows():
    print(index)
    print(row['Units Sold'])
    print("\n")
```
